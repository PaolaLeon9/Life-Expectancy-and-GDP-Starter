# Life Expectancy and GDP 

This project will investigate if there is a strong correlation between the economic output of a country and the life expectancy of its citizens.

Here are a few questions that this project will seek to answer:

* Has life expectancy increased over time in the six nations?
* Has GDP increased over time in the six nations?
* Is there a correlation between GDP and life expectancy of a country?
* What is the average life expectancy in these nations?
* What is the distribution of that life expectancy?

_Data sources_

GDP Source: World Bank national accounts data, and OECD National Accounts data files.

Life expectancy Data Source: World Health Organization

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Summary statistics and examination: 

```python
df = pd.read_csv('all_data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

2. Graphs to explain data

```python
sns.barplot(data=df, x='Year', y='Life expectancy at birth (years)', hue='Country')
plt.title('Life expectancy per year')
plt.xticks(rotation=90)
plt.show()
```

In the previous chart, we can see how life expectancy increase by year as well as in the next graph. 

```python
sns.lineplot(data=df, x='Year', y='Life expectancy at birth (years)', hue=df.Country, style=df.Country, markers=False, dashes=False)
plt.title('Life expectancy between 2000 and 2014')
plt.xticks(rotation=90)
plt.legend(loc=6)
plt.show()
```

Plotting considering country, we can see there isn't big difference in the increment of life expectancy from 2000 to 2014 between United States of America, Mexico, Germany, China and Chile. But, there is a remarkable difference in Zimbabwe.

Now, let's see the changes in GDP.  

```python
sns.boxplot(data=df, x='Year', y='GDP')
plt.xticks(rotation=90)
plt.title('Data distribution of GDP per year')
plt.show()
```

```python
ax=sns.lineplot(data=df, x='Year', y='GDP', hue='Country', alpha=0.6)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.title('GDP per Country between 2000 and 2014')
plt.show()
```

This graph show us the behavior of GDP over the years. We note is stagnat for three countries: Zimbabwe, Chile and Mexico. In the Case of Germany there are small variations, and we can see a remarkable increase for China and United States of America. 

Let's plot GDP vs Life Expectancy for each country.

```python
g = sns.FacetGrid(data=df, col='Country', hue='Country', col_wrap=3, sharey = False, sharex = False)
g.map(sns.scatterplot, 'Life expectancy at birth (years)', 'GDP')
```

In most of the cases, life expectancy increases as GDP. 

## Conclusions

This graphs show us there is a relationship between Life expectancy and GDP because we can note Life expectancy increases as GDP does.
